```python
import sys

def check_positive_number(x, y):
    return x >= 0 and y >= 0 and (x != 0 or y != 0)

def generate_prime_numbers(s, e):
    if not check_positive_number(s, e):
        print("\nInvalid input. Start and end must be non-negative and not both zero.")
        return

    print(f"\nPrime numbers between {s} and {e} are:", end=" ")
    for i in range(s, e + 1):
        print(((i*i)+i+41), end=" ")
    print()

def main():
    count_for_choice = 0
    while True:
        print("\nEnter the choice:")
        print("\n1. Generate prime numbers.")
        print("2. Exit the program.\n")
        try:
            choice = int(input("your choice: "))
        except ValueError:
            print("\nInvalid input. Please enter a number.")
            count_for_choice += 1
            if count_for_choice >= 3:
                print("Too many incorrect inputs. Exiting...\n\nThank You for using The Program")
                sys.exit()
            continue

        if choice == 1:
            count_for_input = 0
            while True:
                try:
                    s = int(input("\nEnter the starting point: "))
                    e = int(input("Enter the ending point: "))
                except ValueError:
                    print("\nInvalid input. Please enter numbers.")
                    count_for_input += 1
                    if count_for_input >= 3:
                        print("Too many incorrect inputs. Exiting...\n\nThank You for using The Program")
                        sys.exit()
                    continue

                generate_prime_numbers(s, e)

                try:
                    choice_2 = int(input("\nEnter your choice:\n\n1. Continue generating prime numbers.\n2. Exit the program.\n\nyour choice: "))
                except ValueError:
                    print("\nInvalid input. Please enter a number.")
                    count_for_choice += 1
                    if count_for_choice >= 3:
                        print("Too many incorrect inputs. Exiting...\n\nThank You for using The Program")
                        sys.exit()
                    continue

                if choice_2 == 1:
                    continue
                elif choice_2 == 2:
                    print("\nThank You for using The Program")
                    sys.exit()
                else:
                    print("\nWrong input. Please enter 1 or 2.")
                    count_for_choice += 1
                    if count_for_choice >= 3:
                        print("Too many incorrect inputs. Exiting...\n\nThank You for using The Program")
                        sys.exit()

        elif choice == 2:
            print("\nThank You for using The Program")
            sys.exit()
        else:
            print("\nWrong input. Please enter 1 or 2.")
            count_for_choice += 1
            if count_for_choice >= 3:
                print("Too many incorrect inputs. Exiting...\n\nThank You for using The Program")
                sys.exit()

if __name__ == "__main__":
    main()
```
